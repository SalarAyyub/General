```python
#Question 2
#ANSWER:
#Given that Business_IDs are unique but may represent same business (chain). 
#So we proceed by Business_Name anyways.

data = pd.DataFrame(dataset)

most_reviewed_business = data.groupby('Business_Name')['Avg_Business_Star_Rating'].count().idxmax()
most_reviews = data['Avg_Business_Star_Rating'][most_reviewed_business].count()

print(f"The business with the highest number of reviews is: {most_reviewed_business}")
print(f"The total number of reviews is: {most_reviews}")
```

```python
#Query returns a business called 'Hash House A Go Go' but there is a type error apparently. 
#So we use isinstance to see if it is what it seems like.

max_rated_business = data.groupby('Business_Name')['Avg_Business_Star_Rating'].count().idxmax()

# Handle potential issues
if isinstance(max_rated_business, str) and max_rated_business in data['Business_Name'].unique():
    most_reviews = data[data['Business_Name'] == max_rated_business]['Avg_Business_Star_Rating'].count()
    print(f"The business with the highest number of reviews is: {max_rated_business}")
    print(f"The total number of reviews is: {most_reviews}")
else:
    print("Business not found or data inconsistency.")

#Looks like there was a type error but we have found the business and number of reviews in garnered. 
```

```python
#Question 3
#ANSWER
#Given that targeted state for minimum 1 5.0 reviw is Nevada (NV)

targeted_state = 'NV'

filtered_data = data[(data['State'] == targeted_state) & (data['Avg_Business_Star_Rating'] == 5.0)]

city_name = filtered_data['City'].unique()

print(f"City names with atleast one 5.0 review are: {city_name}")
```

```python
#Question 4
#ANSWER:
#Given that our targeted category is “Hotels & Travel”. 
#So we query and find that Las Vegas has the highest numver of reviews at 10,245 reviews.

given_category = 'Hotels & Travel'

filtered_data = data[data['Business_Category'] == given_category]
review_counts = filtered_data.groupby('City')['Avg_Business_Star_Rating'].count()

most_reviewed_city = review_counts.idxmax()
highest_count = review_counts.max()

print(f"The city with the highest number of reviews in the '{given_category}' category is: {most_reviewed_city}")
print(f"The total number of reviews is: {highest_count}")
```

```python
#Question 5
#ANSWER: 
#Since "Review_Date" is boject, lets convert it to date time
#so that we can extract the day from it and find the week day. 
#So we find that Monday is usually the day to get reviewed.

data['Review_Date'] = pd.to_datetime(data['Review_Date'], format = '%Y-%m-%d')
data['Day of the week'] = data['Review_Date'].dt.day_name() #makes a new column with day name orresponding to the day in the Review_Date

review_counts = data.groupby('Day of the week').size()
likely_day = review_counts.idxmax()

print(f"Likely day to be reviewed is: {likely_day}")
```

```python
#Question 6
#ANSWER
#We use Bokeh for visuals since we have a vast dataset with 227581 rows.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
```

```python
data['Review_Date'] = pd.to_datetime(data['Review_Date'], format = '%Y-%m-%d')

grouped_data = data.groupby(['Business_Name', pd.Grouper(key='Review_Date', freq='ME')])['Avg_Business_Star_Rating'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Review_Date', y='Avg_Business_Star_Rating', hue='Business_Name', data=grouped_data, palette='viridis', linewidth=1, marker='o')
plt.grid(True)
plt.title('Average Performance Over Time by Business')
plt.xlabel('Review Date')
plt.ylabel('Average Rating')
plt.legend(title='Business Name')
plt.savefig('average_rating_trend.png', dpi=150)
plt.show()

#This part ran gave us the image that we have already pasted as the first image under Question 6. 
```
